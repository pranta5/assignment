Q1. How can you create a Bokeh plot using Python code?

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure,output_file , show
```

```python
from bokeh.sampledata.iris import flowers
```

```python
flowers.head()
```

```python
output_file("plot.html")
p = figure(title='My Bokeh Plot', x_axis_label='X Axis', y_axis_label='Y Axis')
p.circle(flowers['petal_length'],flowers['petal_width'])
show(p)
```

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Glyphs in Bokeh are visual shapes that can be added to a plot to represent data points, lines, or areas. They are the building blocks of a Bokeh plot and can be used to customize the appearance of the data.

```python
output_file('glyphs.html')
p1 = figure(title='glyphs Plot', x_axis_label='X Axis', y_axis_label='Y Axis')
p1.circle(flowers['petal_length'],flowers['petal_width'],size=10)  # add a circle glyph to the figure
show(p1)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python

output_file('legend.html')
p2 = figure(title='legend Plot', x_axis_label='X Axis', y_axis_label='Y Axis')
p2.circle(flowers['petal_length'],flowers['petal_width'],size=10 ,fill_color= 'red',legend_label='Red points') 
p2.legend.location = 'top_left'
p2.title.text_font_style = 'italic'
p2.xaxis.axis_label_text_font_size= '16pt'

show(p2)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

Bokeh server is a component of Bokeh that enables you to build interactive web applications that can be served up in a browser.

To create a Bokeh server application, you typically start by defining your plot or visualization using Bokeh's high-level plotting API. You can then define the interactive behavior you want to add to the plot, such as sliders, buttons, or dropdowns, using Bokeh's interactive widgets.

Once you have defined your plot and interactive widgets, you can then use Bokeh's server-side API to create a Python function that defines how the plot should update in response to user input. This function can use the data or user input to update the plot, and then send the updated plot back to the browser.

To run your Bokeh server application, you simply start the Bokeh server and point your browser to the appropriate URL. The Bokeh server will then serve up your application, allowing users to interact with the plot in real-time.
