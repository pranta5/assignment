### Q2.What is the difference between Bernoulli Naive Bayes and Multinomial Naive Bayes?

- Bernoulli Naive Bayes:

Assumes binary features: Each feature can only take on two values, typically represented by 0 and 1.

Treats features as presence/absence indicators: A value of 1 indicates the feature is present, while 0 indicates it's absent.
Ignores the frequency or count of the features.

Used when features are naturally binary or binarized (e.g., text classification by word presence/absence).

- Multinomial Naive Bayes:

Assumes discrete features: Features can take on a finite number of discrete values, but not limited to just 0 and 1.

Considers the frequency or count of each feature value.

Useful for cases where feature occurrence matters, like word counts in documents, or presence of different events.

## Q3. How does Bernoulli Naive Bayes handle missing values?

- Imputation:

You might choose to impute missing values with the most common value (mode) of the feature. If 0 is more common, you could impute missing values as 0; if 1 is more common, you could impute missing values as 1.

Another option is to use a more sophisticated imputation technique, such as k-nearest neighbors imputation, to estimate the missing values based on other similar instances.

## Q4. Can Gaussian Naive Bayes be used for multi-class classification?

Yes, Gaussian Naive Bayes can be used for multi-class classification. Gaussian Naive Bayes is an extension of the Naive Bayes algorithm that assumes that continuous features follow a Gaussian (normal) distribution within each class. While it is often used for binary classification problems, it can also be extended to handle multiple classes.

```python
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
```

```python
bernoulli_nb = BernoulliNB()
multinomial_nb = MultinomialNB()
gaussian_nb = GaussianNB()
```

```python
# Define 10-fold cross-validation

cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
```

```python
# Evaluate classifiers
def evaluate_classifier(classifier, X, y):
    accuracy = cross_val_score(classifier, X, y, cv=cv, scoring='accuracy')
    precision = cross_val_score(classifier, X, y, cv=cv, scoring='precision_macro')
    recall = cross_val_score(classifier, X, y, cv=cv, scoring='recall_macro')
    f1 = cross_val_score(classifier, X, y, cv=cv, scoring='f1_macro')
    return accuracy, precision, recall, f1
```

```python
# Evaluate Bernoulli Naive Bayes
accuracy_b, precision_b, recall_b, f1_b = evaluate_classifier(bernoulli_nb, X, y)

# Evaluate Multinomial Naive Bayes
accuracy_m, precision_m, recall_m, f1_m = evaluate_classifier(multinomial_nb, X, y)

# Evaluate Gaussian Naive Bayes
accuracy_g, precision_g, recall_g, f1_g = evaluate_classifier(gaussian_nb, X, y)
```

```python
# Print results
print("Bernoulli Naive Bayes:")
print(f"Accuracy: {accuracy_b.mean():.4f}")
print(f"Precision: {precision_b.mean():.4f}")
print(f"Recall: {recall_b.mean():.4f}")
print(f"F1 Score: {f1_b.mean():.4f}")
print("\n")

print("Multinomial Naive Bayes:")
print(f"Accuracy: {accuracy_m.mean():.4f}")
print(f"Precision: {precision_m.mean():.4f}")
print(f"Recall: {recall_m.mean():.4f}")
print(f"F1 Score: {f1_m.mean():.4f}")
print("\n")

print("Gaussian Naive Bayes:")
print(f"Accuracy: {accuracy_g.mean():.4f}")
print(f"Precision: {precision_g.mean():.4f}")
print(f"Recall: {recall_g.mean():.4f}")
print(f"F1 Score: {f1_g.mean():.4f}")
```

Difficulty Handling Missing Values: Naive Bayes doesn't handle missing values well, and appropriate preprocessing or imputation is required.
