```python
The for loop is used when we already know the number of iterations
```

```python
When we need to end the loop on a condition other than the number of times, we use a while loop.
```

```python
for i in range(0,6):
    print(i)
```

```python
i=0
while i<6:
    print(i)
    i=i+1
```

```python
#2
x=0
m=1
for i in range(1,11):
    x=x+i
    m=m*i
    i=i+1
print("sum:",x)
print("product:",m)
```

```python
#3
units=float(input("Enter units: "))
if units<=100:
    print("bill:",units*4.5)
elif units>100 and units <=200:
    print("bill:",(units-100)*6+450)
elif units>200 and units<=300:
    print("bill:",(units-200)*10+450+600)
elif units>300:
    print("bill:",(units-300)*20+450+600+1000)
    
```

```python
#4
l1=list(range(1,101))
l2=[]
for i in l1:
    if i/4 and i/5:
        l2.append(i)
    print(l2)
```

```python
lst=[]
l5=[]
for i in range(1,101):
    lst.append(i)
    for j in lst:
        if j/4 and j/5:
            l5.append(j)
print(l5)
    
```
